```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Question 1: Simple Line Plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.title('Line Plot of x vs y')
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
# Question 2: Scatter Plot
x_scatter = np.random.randint(1, 100, 50)
y_scatter = np.random.randint(1, 100, 50)

plt.scatter(x_scatter, y_scatter)
plt.title('Scatter Plot of Random Data')
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
# Question 3: Bar Plot
students = ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve']
scores = [88, 92, 79, 85, 90]

plt.bar(students, scores)
plt.title('Student Test Scores')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.show()
```

```python
# Question 4: Histogram
test_scores = np.random.randint(50, 101, 100)

plt.hist(test_scores, bins=8)
plt.title('Histogram of Test Scores')
plt.xlabel('Score Range')
plt.ylabel('Frequency')
plt.show()
```

```python
# Question 5: Customizing Line Plot from Question 1
plt.plot(x, y, 'r--')  # Red dashed line
plt.title('Customized Line Plot')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.annotate('Peak', xy=(5, 25), xytext=(3, 20),
             arrowprops=dict(facecolor='black'))
plt.show()
```

```python
# Question 6: Box Plot
class1_scores = np.random.randint(60, 101, 50)
class2_scores = np.random.randint(60, 101, 50)

plt.boxplot([class1_scores, class2_scores], labels=['Class 1', 'Class 2'])
plt.title('Box Plot of Class Scores')
plt.xlabel('Class')
plt.ylabel('Scores')
plt.show()
```

```python
# Question 7: Stacked Bar Plot for Product Sales by Salesperson
salespeople = ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve']
product_A = [30, 25, 35, 20, 30]
product_B = [20, 30, 15, 25, 20]
product_C = [25, 15, 20, 30, 25]

plt.bar(salespeople, product_A, label='Product A')
plt.bar(salespeople, product_B, bottom=product_A, label='Product B')
plt.bar(salespeople, product_C, bottom=np.add(product_A, product_B), label='Product C')
plt.title('Sales by Product and Salesperson')
plt.xlabel('Salesperson')
plt.ylabel('Units Sold')
plt.legend()
plt.show()
```
